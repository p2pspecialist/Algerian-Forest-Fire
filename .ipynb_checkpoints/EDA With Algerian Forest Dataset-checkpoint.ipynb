```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_excel('Algerian_forest_fires_dataset_UPDATE.xlsx', header=1)
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
```

```python
dataset.head()
```

```python
df = dataset
```

```python
df.head()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.tail()
```

```python
# Remove null values
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# Remove 122 row as it contains same header
df=df.drop(122).reset_index(drop=True)
```

```python
# Fix extra spaces in column headers

df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

```python
# Change the data type of the columns to integer and float as per data available in column

df[['day','month','year','Temperature','RH','Ws']]=df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df[['Rain','FFMC','DMC','DC','ISI','BUI','FWI']]=df[['Rain','FFMC','DMC','DC','ISI','BUI','FWI']].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
# Now lets save the cleaned up file into csv format for futher use

df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index=False)
```

# Exploratory Data Analysis

```python
# Dropp Day, month and year column as we need to focus on other columns

df_copy = df.drop(['day','month','year'], axis=1)
```

```python
df_copy.head()
```

```python
# Categorical features values
df_copy['Classes'].value_counts()
```

```python
# Encoding of the categorical features
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy.tail()
```

```python
# Plot density for all features
plt.style.use('seaborn')
df_copy.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
# Percentage for pie chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
# Plotting pie chart
classlabels = ['Not Fire', 'Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage, labels=classlabels,autopct='%1.1f%%')
plt.title('Pie Chart of Classes')
plt.show()
```
